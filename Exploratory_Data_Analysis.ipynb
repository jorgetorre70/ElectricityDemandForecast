```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot

jtplot.style(theme='grade3', context='notebook', ticks=True, grid=False)
sns.set_theme()
```

Importing data from selected locations

```python
demandaMTY = pd.read_csv('df_fullMTY.csv')
demandaVDMNORTE = pd.read_csv('df_fullVDMNORTE.csv')
demandaVDMSUR = pd.read_csv('df_fullVDMSUR.csv')
demandaVDMCENTRO = pd.read_csv('df_fullVDMCENTRO.csv')
demandaGDL = pd.read_csv('df_fullGDL.csv')
demandaPUEBLA = pd.read_csv('df_fullPUEBLA.csv')
```

```python
demandaMTY['Year'] = pd.to_datetime(demandaMTY['Fecha'])
demandaMTY.set_index('Year', inplace=True)
demandaMTY = demandaMTY.loc[(demandaMTY.index >= '2017-01-01') & (demandaMTY.index <= '2023-12-31')]
demandaMTY = demandaMTY.drop(columns=['Fecha'])

demandaVDMNORTE['Year'] = pd.to_datetime(demandaVDMNORTE['Fecha'])
demandaVDMNORTE.set_index('Year', inplace=True)
demandaVDMNORTE = demandaVDMNORTE.loc[(demandaVDMNORTE.index >= '2017-01-01') & (demandaVDMNORTE.index <= '2023-12-31')]
demandaVDMNORTE = demandaVDMNORTE.drop(columns=['Fecha'])

demandaVDMSUR['Year'] = pd.to_datetime(demandaVDMSUR['Fecha'])
demandaVDMSUR.set_index('Year', inplace=True)
demandaVDMSUR = demandaVDMSUR.loc[(demandaVDMSUR.index >= '2017-01-01') & (demandaVDMSUR.index <= '2023-12-31')]
demandaVDMSUR = demandaVDMSUR.drop(columns=['Fecha'])

demandaVDMCENTRO['Year'] = pd.to_datetime(demandaVDMCENTRO['Fecha'])
demandaVDMCENTRO.set_index('Year', inplace=True)
demandaVDMCENTRO = demandaVDMCENTRO.loc[(demandaMTY.index >= '2017-01-01') & (demandaVDMCENTRO.index <= '2023-12-31')]
demandaVDMCENTRO = demandaVDMCENTRO.drop(columns=['Fecha'])

demandaGDL['Year'] = pd.to_datetime(demandaGDL['Fecha'])
demandaGDL.set_index('Year', inplace=True)
demandaGDL = demandaGDL.loc[(demandaMTY.index >= '2017-01-01') & (demandaGDL.index <= '2023-12-31')]
demandaGDL = demandaGDL.drop(columns=['Fecha'])

demandaPUEBLA['Year'] = pd.to_datetime(demandaPUEBLA['Fecha'])
demandaPUEBLA.set_index('Year', inplace=True)
demandaPUEBLA = demandaPUEBLA.loc[(demandaMTY.index >= '2017-01-01') & (demandaPUEBLA.index <= '2023-12-31')]
demandaPUEBLA = demandaPUEBLA.drop(columns=['Fecha'])
```

Plotting the data for comparison

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

sns.lineplot(ax=axes[0, 0], data=demandaMTY, legend=False)
axes[0, 0].set_title('MONTERREY')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Energy (MWh)')

sns.lineplot(ax=axes[0, 1], data=demandaGDL, legend=False)
axes[0, 1].set_title('GUADALAJARA')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Energy (MWh)')

sns.lineplot(ax=axes[0, 2], data=demandaVDMNORTE, legend=False)
axes[0, 2].set_title('VDM NORTE')
axes[0, 2].set_xlabel('Year')
axes[0, 2].set_ylabel('Energy (MWh)')

sns.lineplot(ax=axes[1, 0], data=demandaVDMCENTRO, legend=False)
axes[1, 0].set_title('VDMCENTRO')
axes[1, 0].set_xlabel('Year')
axes[1, 0].set_ylabel('Energy (MWh)')

sns.lineplot(ax=axes[1, 1], data=demandaVDMSUR, legend=False)
axes[1, 1].set_title('VDMSUR')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Energy (MWh)')

sns.lineplot(ax=axes[1, 2], data=demandaPUEBLA, legend=False)
axes[1, 2].set_title('PUEBLA')
axes[1, 2].set_xlabel('Year')
axes[1, 2].set_ylabel('Energy (MWh)')

plt.tight_layout()
plt.show()
fig.savefig('figure4.jpg', format='jpg')
```
